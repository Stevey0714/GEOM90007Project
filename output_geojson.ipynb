```python
import os
import pandas as pd
import json
```

```python
restaurant_output = {"type": "geojson", "data": {"type": "FeatureCollection", "features": []}}
hotel_output = {"type": "geojson", "data": {"type": "FeatureCollection", "features": []}}
cafe_output = {"type": "geojson", "data": {"type": "FeatureCollection", "features": []}}
```

```python
rest_df = pd.read_csv('data_restaurant.csv')
hotel_df = pd.read_csv('data_hotel.csv')
cafe_df = pd.read_csv('data_cafe.csv')
```

```python
cafe_df
```

```python
rest_features = []
for index, row in rest_df.iterrows():
    feature = {'type': 'Feature', 'geometry': {'type': 'Point', 'coordinates': []}, 'properties': {}}
    feature['properties']['rank'] = int(row['Rank'])
    feature['properties']['name'] = row['Name']
    feature['properties']['address'] = row['Address']
    feature['properties']['price'] = int(row['Price'])
    feature['properties']['rating'] = float(row['Rating'])
    feature['properties']['cuisine'] = sorted(row['Cuisine'].split(", "))
    feature['geometry']['coordinates'] = [float(coord) for coord in row['Coordinate'].split(", ")][::-1]
    rest_features.append(feature)
restaurant_output['data']['features'] = rest_features
```

```python
with open('rest_geojson.geojson', 'w+', encoding='utf-8') as f:
    json.dump(restaurant_output, f, ensure_ascii=False, indent=4)
```
